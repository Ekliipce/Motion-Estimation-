### Estimation de mouvement par une methode pixel-wise 

Avec admission de l’hypothese de l’intensite constante, nous pouvons utiliser l'expression de l'EDFD pour déterminer les vecteurs de mouvements des pixels de l'image 1 vers ceux de l'image 2.

On déterminera un vecteur de mouvement pour chaque pixels



```python
def criterionRGB(F1, F2):
    """
    Transforme les matrices F1, F2 de shape (h, w, c) en matrice (h, w) selon le critère
    Ici nous ferons la somme des 3 canaux rgb
    
    F1, F2 : nd.array de shape(h, w, c)
    return f1, f2 : nd.array de shape (h, w)
    """
    
    h, w = F1.shape[0], F1.shape[1]
    f1 = np.array([[F1[i, j].sum() for j in range(w)] for i in range(h)], dtype='int')
    f2 = np.array([[F2[i, j].sum() for j in range(w)] for i in range(h)], dtype='int') 
    
    return f1, f2
```

```python
print(f1.shape)
a, b = criterionRGB(f1, f2)
a.shape
```

```python
def DFD_pixelwise(img1, img2, x, y, r):
    """
    img1, img2 : nd.array de shape (h, w)
    x, y : coordonné x, y du pixel courant
    r: voisinage du pixel courrant
    
    return a_x, a_y : Coordonnées x et y du vecteur d, minimisant l'EDFD 
    """
    a_x, a_y = 0, 0
    dfd_min = np.inf
    
    ## Ici un parcours spécial de -r à r selon les axes y puis x
    ## Ce parcours fait passer les pixels les plus proches du pixel (x, y) en premier
    
    for i in np.concatenate([np.arange(0, r), np.arange(-r+1, 0)]):
        for j in np.concatenate([np.arange(0, r), np.arange(-r+1, 0)]):
            dfd = np.abs(img2[y+i, x+j] - img1[y, x])

            if (dfd < dfd_min):
                dfd_min = dfd
                a_y, a_x = i, j
                
    return a_x, a_y
```

```python
def DFD_pixelwise2(F1, F2, x, y, r):
    Window_f1 = np.full((2*r+1, 2*r+1), F1[y, x].sum())
    Window_f2 = F2[y-r:y+r+1, x-r: x+r+1]
    
    diff = np.abs(Window_f2 - Window_f1)
    miny, minx = np.unravel_index(np.argmin(diff, axis=None), diff.shape)
    return -miny, -minx
```

```python
def computeVector(r, p, function):
    """
    Renvoie la liste de chaque vecteurs correspondant à chaque pixel
    
    r: voisinage de chaque pixel
    p: pas de calcul entre chaque pixel
    function: fonction de calcule du vecteur
    
    return V1 : la liste des coordonnées x de chaque vecteurs de mouvements
           V2 : la liste des coordonnées y de chaque vecteurs de mouvements
    """
    V1, V2 = [], []
    
    for i in range(r, h1-r, p):
        for j in range(r, w1-r, p):
            v1, v2 = function(f1_grey, f2_grey, j, i, r)
            V1.append(v1)
            V2.append(v2)
    return V1, V2
```

```python
%%time
r = 5
p = 7
V1, V2 = computeVector(r, p, DFD_pixelwise)
```

```python
def displayVector(V1, V2, r, pas, w1, h1, img):
    """
    Affichage des vecteurs sur l'image img
    """
    x = np.arange(r, w1-r, pas)
    y = np.arange(r, h1-r, pas)
    X, Y = np.meshgrid(x, y)

    sh1, sh2 = X.shape[0], X.shape[1]
    fig, ax = plt.subplots(figsize =(10, 7))
    plt.imshow(img)
    ax.quiver(X, Y, np.array(V1, dtype=float).reshape((sh2, sh1)), np.array(V2, dtype=float).reshape((sh2, sh1)))
    plt.show()
```

```python
displayVector(V1, V2, r, p, w1, h1, F1)
```

### Estimation de mouvement par blocs (BMA) pour différentes tailles de blocs.

Toujours avec admission de l’hypothese de l’intensite constante, on pourra utiliser une méthode exhautive pour déterminer le mouvments des blocs.

```python
def DFD_blockWise(img1, img2, x, y, r):
    """
    img1, img2 : nd.array de shape (h, w)
    x, y : coordonné x, y du centre du bloc
    r: rayon du bloc
    
    return a_x, a_y : Coordonnées x et y du vecteur minimisant l'EDFD du bloc courant
    """
    a_x, a_y = 0, 0
    dfd_min = np.inf
    
    ## Le parcours de -2r à 2r passera tous les blocs autours du blocs courant
    ## Avec un pas de r/2, on aura 16 blocs de comparaisons
    ## On choisira celui qui minisise l'EDFD pour déterminer (a_x, a_y)
    
    for i in np.concatenate([np.arange(0, 2*r), np.arange(-2*(r+1), 0)])[::int(r/2)]:
        for j in np.concatenate([np.arange(0, 2*r), np.arange(-2*(r+1), 0)])[::int(r/2)]:
            
            dfd = 0
            for t1 in np.concatenate([np.arange(0, r), np.arange(-r+1, 0)]):
                for t2 in np.concatenate([np.arange(0, r), np.arange(-r+1, 0)]):
                    if (y + i + t1 > 0 and y + i + t1 < img1.shape[0] and
                        x + j + t2 > 0 and x + j + t2 < img1.shape[1]):
                        dfd += np.abs(img2[y+i+t1, x+j+t2] - img1[y+t1, x+t2])

            if (dfd < dfd_min):
                dfd_min = dfd
                a_y, a_x = i, j
    return a_y, a_y
```

```python
%%time
r = 5
p = 10
V1B, V2B = computeVector(r, p, DFD_blockWise)
displayVector(V1B, V2B, r, p, w1, h1, F1)
```

### Estimation de mouvement en Node Based

On prend maintenant en compte le fait que les objets peuvent subir des déformations (zooms, translations, rotations). On calculera ces déformations pour estimer le mouvement.

On ne pourra pas utiliser de méthodes exhaustives comme pour les deux premières méthodes. On utilisera alors une descente de gradient pour estimer le vecteur qui minimise l'EDFD

On se donne ici, deux images plus simples pour le problème :

```python
img2_F1 = Image.open('./data/1.png', mode='r', formats=None).convert('L')
img2_F2 = Image.open('./data/2.png', mode='r', formats=None).convert('L')

plt.figure(figsize=(10, 10))
plt.subplot(1, 2, 1)
plt.imshow(img2_F1)
plt.subplot(1, 2, 2)
plt.imshow(img2_F2)
```

##### Image 1 décomposée en block

```python
def nodalOneBlock(x, y, r):
    return (x-r, y-r), (x+r, y-r), (x+r, y+r), (x-r, y+r)
```

```python
r = int(img2_h1 / 4 / 2)
r
```

```python
def displayNodal(r, h1, w1, fnct):
    """
    Affiche l'image avec tous les noeuds des blocs, détermininés par leur rayon r
    """
    
    for i in range(r, h1, r*2):
        for j in range(r, w1, r*2):
            d1, d2, d3, d4 = fnct(j, i, r)
            sqx = [d1[0], d2[0], d3[0], d4[0], d1[0]]
            sqy = [d1[1], d2[1], d3[1], d4[1], d1[1]]
            plt.plot(sqx, sqy, "-")
    plt.imshow(img2_F1)
    plt.show()
```

```python
displayNodal(r, img2_h1, img2_w1, nodalOneBlock)
```

##### Image 2 montrant les déformations effectuées sur l'Image 1

```python
def EDFD_node_based(list_d, img1, img2, x, y, r):
    """
    Calcule de l'edfd pour node base entre l'image 1 et 2
    
    img1, img2: nd.array de shape (h, w)
    x, y : coordonnées x, y du centre du bloc
    r : rayon du bloc
    list_d : liste des paramètres noeuds à optimiser
    
    return : edfd entre img1, img2 
    """
    xmax, xmin, ymax, ymin = x+r, x-r, y-r, y+r
    cnst = (xmax - xmin) * (ymax - ymin)
    
    d1 = np.array([list_d[0], list_d[1]])
    d2 = np.array([list_d[2], list_d[3]])
    d3 = np.array([list_d[4], list_d[5]])
    d4 = np.array([list_d[6], list_d[7]])
    #print(d1, d2, d3, d4)
    
    dfd = 0
    pas = 2
    
    for i in np.arange(-r, r+1)[::pas]:
        for j in np.arange(-r, r+1)[::pas]:
            curr_x = x + j
            curr_y = y + i 
            
            
            ## Calcule de d selon les polynomes d'interpolations
            d = 0
            d += (xmax - curr_x) * (ymax - curr_y) * d1 / cnst 
            d += (curr_x - xmin) * (ymax - curr_y) * d2 / cnst
            d += (xmax - curr_x) * (curr_y - ymin) * d3 / cnst
            d += (curr_x - xmin) * (curr_y - ymin) * d4 / cnst
            d = np.ceil(d).astype('int64')
            
            if (y + i + d[1] > 0 and y + i + d[1] < img1.shape[0] and
                x + j + d[0] > 0 and x + j + d[0] < img1.shape[1]):
                dfd += ((img2[y + i + d[1], x + j + d[0]] - img1[y + i, x + j])**2)*pas
            
    
    return dfd
```

```python
def NodeBased(x, y, r, img1 = img2_f1, img2 = img2_f2):
    """
    Calcule la place supposée des noeuds par la déformation de l'imgage 1 à l'image 2
    
    img1, img2: nd.array de shape (h, w)
    x, y : coordonnées x, y du centre du bloc
    r : rayon du bloc
    
    return : les coins du blocs déplacés par la déformation   
    """
    ld = [0] * 8
    gamma = 1e-5
    epochs = 10
    h = 1
    
    old_ld = []
    
    # Descente de Gradient
    for epoch in range(1, epochs+1):
        
        ## Early stopping 
        if (epoch >= 2 and ld == old_ld):
            break
        old_ld = ld.copy()
        
        ## Ajustment des parmaramètres l
        for param_l, l in zip(ld, range(0,8)):
            ld2 = ld.copy()
            ld2[l] += h
            
            EDFD = EDFD_node_based(ld, img1, img2, x, y, r) 
            EDFDh = EDFD_node_based(ld2, img1, img2, x, y, r)
            
            
            partial_derivative = (EDFDh - EDFD) / h
            
            param_l -= gamma * partial_derivative
            ld[l] = param_l
        
    return (x-r + ld[4], y-r + ld[5]), (x+r + ld[6], y-r + ld[7]), \
           (x+r + ld[2], y+r + ld[3]), (x-r + ld[0], y+r + ld[1])
```

```python
%%time
displayNodal(r, img2_h1, img2_w1, NodeBased)
```
